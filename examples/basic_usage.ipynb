# Inference Library - Basic Usage

This notebook demonstrates how to use the inference library for LLM research.

## Setup

First, make sure you've copied `.env.example` to `.env` and added your API keys.

```python
# Import the public API
from inference import create_client, run_completion, InferenceRequest, InferenceResult
```

## Single Completion

Run a single inference request:

```python
import asyncio

async def single_completion():
    # Create client from config
    client = create_client("config/inference.example.yaml")
    
    # Create request
    request = InferenceRequest(
        model_alias="mock-test",
        prompt="What is the capital of France?"
    )
    
    # Run completion
    result = await run_completion(client, request)
    
    # Display result
    print(f"Provider: {result.provider}")
    print(f"Model: {result.model}")
    print(f"Content: {result.content}")
    print(f"Tokens: prompt={result.prompt_tokens}, completion={result.completion_tokens}")
    print(f"Latency: {result.latency_ms:.1f} ms")

await single_completion()
```

## Batch Processing

Process multiple requests with checkpoint/resume:

```python
from inference import run_batch

async def generate_requests():
    """Generate a stream of inference requests."""
    prompts = [
        "What is 2+2?",
        "What is the capital of Germany?",
        "Name a primary color."
    ]
    
    for prompt in prompts:
        yield InferenceRequest(
            model_alias="mock-test",
            prompt=prompt
        )

async def run_batch_example():
    print("Running batch inference...")
    await run_batch("config/inference.example.yaml", generate_requests())
    print("Batch completed! Check logs/inference.jsonl for results.")

await run_batch_example()
```
